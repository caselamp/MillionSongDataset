```python
import pandas as pd
```

```python
df = pd.read_csv('./big_subset.csv')
#df["artist_terms_array"] = ""
```

```python
df.head()
```

```python
words = df['artist_terms'][0].split("\n")
descr = []
for word in words:
    wo = word.split("' '")
    for w in wo:
        descr.append(w)
zmod = [zz.replace('\'', '').replace('[', '').replace(']', '').strip() for zz in descr]
```

```python
df.at[0,'artist_terms'] = zmod
#z_string = ', '.join(zmod)
#df.at[0,'artist_terms'] = z_string
```

```python
len(df)
```

```python
for i in range(len(df)):
    words = df['artist_terms'][i].split("\n")
    descr = []
    for word in words:
        wo = word.split("' '")
        for w in wo:
            descr.append(w)
    zmod = [zz.replace('\'', '').replace('[', '').replace(']', '').strip() for zz in descr]
    df.at[i,'artist_terms'] = zmod
```
